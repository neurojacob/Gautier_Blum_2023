## Integrate neurons from Gautier and Yadav datasets

```r
gautier_neurons <- readRDS(file = '/oak/stanford/groups/agitler/Shared/Matters_Arising/Gautier_Yadav/gautier_neurons.rds')
gautier_neurons$study <- 'Gautier'
```

```r
yadav_neurons <- readRDS(file = '/oak/stanford/groups/agitler/Shared/Matters_Arising/Gautier_Yadav/yadav_neurons.rds')
yadav_neurons$study <- 'Yadav'
```

```r
yadav_neurons.list <- SplitObject(yadav_neurons, split.by = "orig.ident")
combo.list <- c(gautier_neurons, yadav_neurons.list)
```

```r
combo.list
```

```r
# normalize and identify variable features for each dataset independently
combo.list <- lapply(X = combo.list, FUN = function(x) {
    x <- NormalizeData(x)
    x <- FindVariableFeatures(x, selection.method = "vst", nfeatures = 20000)
})
```

```r
# select features that are repeatedly variable across datasets for integration
features <- SelectIntegrationFeatures(object.list = combo.list, nfeatures = 20000)
```

```r
anchors <- FindIntegrationAnchors(object.list = combo.list, anchor.features = features, normalization.method = "LogNormalize", k.filter = 178)
```

```r
# this command creates an 'integrated' data assay
integrated <- IntegrateData(anchorset = anchors, normalization.method = "LogNormalize")
```

```r
# specify that we will perform downstream analysis on the corrected data note that the
# original unmodified data still resides in the 'RNA' assay
DefaultAssay(integrated) <- "integrated"
```

```r
# Run the standard workflow for visualization and clustering
integrated <- ScaleData(integrated, verbose = FALSE)
integrated <- RunPCA(integrated, npcs = 100, verbose = FALSE)
```

```r
saveRDS(integrated, file = '/oak/stanford/groups/agitler/Shared/Matters_Arising/Gautier_Yadav/integrated.rds')
```

```r
# Cluster
neurons.55 <- FindNeighbors(integrated, dims = 1:55)
neurons.55 <- FindClusters(neurons.55, resolution = 0.8)

# UMAP
neurons.55 <- RunUMAP(neurons.55, dims = 1:55)

DimPlot(neurons.55, reduction = "umap", group.by = "study")
DimPlot(neurons.55, reduction = "umap", label = TRUE) 

# Get number of cells per cluster and per study
table(neurons.55@meta.data$seurat_clusters, neurons.55@meta.data$study)
```

```r
c4and8 <- subset(neurons.55, subset = seurat_clusters %in% c(4,8))
```

```r
# Get number of motor and other neurons from each dataset 
table(c4and8@meta.data$motor_neuron, c4and8@meta.data$study)
```

## Subcluster motor neurons (cluster 34)

```r
mns <- subset(neurons.55, subset = integrated_snn_res.0.8 == "34")
DefaultAssay(mns) <- "RNA"
```

```r
mns <- FindVariableFeatures(mns, selection.method = "vst", nfeatures = 28)
all.genes <- rownames(mns)
mns <- ScaleData(mns, features = all.genes)
```

```r
# PCA
mns <- RunPCA(mns, features = VariableFeatures(object = mns), approx = FALSE)
```

```r
# Cluster
mns <- FindNeighbors(mns, dims = 1:2)
mns <- FindClusters(mns, resolution = 0.4, graph.name = "RNA_snn")
```
